```
# import
import requests as req
from bs4 import BeautifulSoup
```

## 2. html을 받아오기


```
# https://finance.naver.com/sise/
url = 'https://finance.naver.com/sise/'
res = req.get(url)
html = res.text
```

```
soup = BeautifulSoup(html,'html.parser')
```

```
s1=soup.find('ul',{'id':'popularItemList'})
# print(s1)
# print(type(s1))
s2=s1.find_all('li')
# print(type(s2))
di = {'기업명':[],'주가':[]}
for s2_ in s2:
#     print(s2_.find('a').text,end=' ')
    di['기업명'].append(s2_.find('a').text)
    
    di['주가'].append( s2_.find('span').text )

```
